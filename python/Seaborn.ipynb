```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

### Correlation with Heatmap
it helps to find inter relation between two independent and dependent feature.

```python
df.dtypes
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

## JoinPlot
A join plot allows to study the relationship between 2 numerical variables.The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or 2D density plot

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.barplot(x="day", y="total_bill", data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
```

```python
sns.histplot(df["total_bill"])
```

```python
sns.kdeplot(df["total_bill"])
```

```python
sns.regplot(x="total_bill", y="tip", data=df)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr)
```
