```python
from io import StringIO, BytesIO
```

```python
data=('col1,col2,col3\n'
      'x,y,2\n'
      'a,b,3\n'
      'd,e,4')
```

```python
type(data)
```

```python
StringIO()
```

```python
pd.read_csv(StringIO(data))
# generally convert comma seperated string data to dataframes without using csv file directly
```

```python
# read from specific columns
df=pd.read_csv(StringIO(data),usecols=['col1','col2'])
```

```python
df
```

```python
# to create a csv file
df.to_csv('test2.csv')
```

```python
data2=('index,a,b,c\n'
      '1,2,3,9\n'
      '4,5,6,7\n')
```

```python
pd.read_csv(StringIO(data2),index_col=2)
```

```python
data3=('a,b,c\n'
      '1,2,3,\n'
      '4,5,6,\n')
```

```python
pd.read_csv(StringIO(data3))
```

```python
pd.read_csv(StringIO(data3),index_col=False)
```

```python
# combining usecols with index_col
pd.read_csv(StringIO(data3),usecols=['b','c'],index_col=False)
```

```python
# use of escaping character in csv file
data4=('a,b\n"hello,\\"Bob\\",nice to see you",5')
```

```python
pd.read_csv(StringIO(data4),escapechar='\\')
# we can also use " sep='' " for \t and other escaping character.
```
