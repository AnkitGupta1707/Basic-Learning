#### How does survival rate vary across different passenger classes (pclass)?

```python
sns.set_style('whitegrid')
s=sns.countplot(x='Survived',hue='Sex',data=train,palette='RdBu_r')
for container in s.containers:
    s.bar_label(container)
plt.xticks([0,1],["died","survived"])
```

The above graph indicates that how many man and women survived and died so basically the amount of male died is far greater than female because as we saw the movie "titanic" in that movie they clearly showed that firstly all the female and kids are tried to be saved and then the male.

## Question 3

### How does age distribution differ between survivors and non-survivors?

```python
sns.histplot(train['Age'].dropna())
```

```python
train["Age"].dropna()
```


### Firstly we drop all the missing values as we saw above that the missing vcalue is present in age column.
### And the above graph shows the count of person in particular age groups.

```python
sns.set_style('whitegrid')
# sns.histplot(hue="Survived",x='Age',data=train)
age=sns.boxplot(x="Survived",y="Age",data=train)
medians = train.groupby('Survived')['Age'].median()
for i, median in enumerate(medians):
    age.text(i, median, f"Median = {median:.1f}",
            horizontalalignment='center',
            verticalalignment='bottom',
            fontweight='bold',
            color='black')
plt.xticks([0,1],["died","survived"])
```

So the above plot shows that the middle age of survivors and non-survivors are same = 28 and average age of survivors and non-survivors are roughly between 20-35 that means young adults

## Question 7

### What is the survival rate for each gender?

#### Male survival percentage

```python
x=(train["Sex"]=='male').sum()
y=((train["Sex"]=='male') & (train["Survived"]==1)).sum()
male_survived= (y/x)*100
male_survived
```

#### Female survival percentage

```python
x1=(train["Sex"]=='female').sum()
y1=((train["Sex"]=='female') & (train["Survived"]==1)).sum()
female_survived= (y1/x1)*100
female_survived
```
