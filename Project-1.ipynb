## Queue Management System

```python
# ===== Interview Queue System =====
status=True
print("""1. Add Interview Candidate
2. Serve Next Candidate
3. Display Waiting Queue
4. Exit""")
candidate_list=[]
debug=0
token_number=1
while(status):
    print("Enter your choice: ")
    choice=int(input())
    if debug:
        print("DEBUG: "+choice)
    if choice==1:
        # 1. Add Interview Candidate
        token="C"+str(token_number)
        candidate_list.append(token)
        token_number+=1
        if debug:
            print("DEBUG: "+"token_number and candidate_list ")
            print("DEBUG: "+token_number,candidate_list)
        print(token)

    elif choice==2:
        # 2. Serve Next Candidate
        print("interviewing candidate : ",candidate_list[0])
        candidate_list.pop(0)
        if debug:
            print("DEBUG: "+"candidate_list")
            print("DEBUG: "+candidate_list)
    elif choice==3:
        # 3. Display Waiting Queue
        for i in range(0,len(candidate_list)):
            print(candidate_list[i],end=" ")
            for j in range(0,i+1):
                print("* ",end="")
            print()
    elif choice==4:
        # 4. Exit
        print("Exiting Interview System...")
        break
    else:
        print("Enter a valid choice")
```
